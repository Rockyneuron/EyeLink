Test for frame rate in pupil adquisition    

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python

data=pd.read_csv('pupil_positions.csv')

ratio_frames=np.diff(data['pupil_timestamp'])
```

For 120Hz the time period of events sould be :1/120=0.008333333333333333

```python
event_time_mean=np.mean(ratio_frames)
event_time_std=np.std(ratio_frames)
```

```python
fig,ax=plt.subplots(1,2)
sns.set_theme()
sns.histplot(data=ratio_frames,bins=100,stat='probability',ax=ax[0])
sns.histplot(data=ratio_frames[ratio_frames>0.03],bins=100,stat='probability',ax=ax[1])
fig.set_figwidth(15)
ax[0].axvline(event_time_mean, color='black',linestyle='dashed')
ax[0].axvline(event_time_mean-event_time_std, color='red',linestyle='dashed')
ax[0].axvline(event_time_mean+event_time_std, color='red',linestyle='dashed')
ax[1].axvline(event_time_mean, color='black',linestyle='dashed')
ax[1].axvline(event_time_mean-event_time_std, color='red',linestyle='dashed')
ax[1].axvline(event_time_mean+event_time_std, color='red',linestyle='dashed')
ax[0].set_title(' mean = {}({}std)'.format(np.round(event_time_mean,3),np.round(event_time_std,3)))
ax[1].set_title(' mean = {}({}std)'.format(np.round(event_time_mean,3),np.round(event_time_std,3)))
```

```python
fig,ax=plt.subplots(1,2)
sns.histplot(data=ratio_frames,bins=100,ax=ax[0])
sns.histplot(data=ratio_frames[ratio_frames>0.02],bins=100,ax=ax[1])
fig.set_figwidth(15)
ax[0].axvline(event_time_mean, color='black',linestyle='dashed')
ax[0].axvline(event_time_mean-event_time_std, color='red',linestyle='dashed')
ax[0].axvline(event_time_mean+event_time_std, color='red',linestyle='dashed')
ax[1].axvline(event_time_mean, color='black',linestyle='dashed')
ax[1].axvline(event_time_mean-event_time_std, color='red',linestyle='dashed')
ax[1].axvline(event_time_mean+event_time_std, color='red',linestyle='dashed')
```

LGTM! Frecuency rate for pupil eye looks good. It is sampling at a higher frecuency
